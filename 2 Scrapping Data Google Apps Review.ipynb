## Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google_play_scraper import Sort, reviews
```

## Mengambil alamat Apps di Google Play Store

Pada notebook ini kita akan mencoba mengambil review dari aplikasi Whatsapp berbahasa Indonesia. Link dari aplikasi yg dibutuhkan oleh API pada library ini adalah karakter tepat setelah id= yaitu **`com.whatsapp`** dan sebelum karakter &hl...

## Scrapping Data Review

```python
result, continuation_token = reviews(
    'com..btpn.dc',                            # Alamat apps
    lang='id',                                 # Bahasa Review , defaults : 'en'
    country='id',                              # Negara Asal Reviewer defaults : 'us'
    sort=Sort.MOST_RELEVANT,                   # Urutan Review yang diambil , defaults : MOST_RELEVANT
    count=1000                                 # Jumlah Review yang diambil, defaults : 100
)
```

```python
result()
```

```python
Hasil_Scrapping = pd.DataFrame(result)
Hasil_Scrapping
```
